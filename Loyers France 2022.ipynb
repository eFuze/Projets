```python
import pandas as pd
import numpy as np
```

```python
loyers = pd.read_csv(("C:/Users/alexi/Downloads/pred-app-mef-dhup.csv"), sep=';', decimal=',')
```

```python
loyers
```

```python
new_column_names = {
'id_zone': 'Identifiant de la zone',
    'INSEE_C': 'Code INSEE',
    'LIBGEO': 'Localisation',
    'EPCI': 'Code EPCI',
    'DEP': 'Département',
    'REG': 'Région',
    'loypredm2': 'Loyer prédit par mètre carré',
    'lwr.IPm2': 'Limite inférieure de l\'IP par mètre carré',
    'upr.IPm2': 'Limite supérieure de l\'IP par mètre carré',
    'TYPPRED': 'Type de prédiction',
    'nbobs_com': 'Nombre d\'observations (Commune)',
    'nbobs_mail': 'Nombre d\'observations (Maille)',
    'R2_adj': 'R² ajusté'
}

loyers1 = loyers.rename(columns=new_column_names)

# Display the renamed DataFrame
print(loyers1)
```

```python
loyers1
```

```python
loyers1 = loyers1[loyers1["R² ajusté"] >= 0.45]
```

```python
# Calculate summary statistics for numerical columns
summary_stats = loyers1.describe()

# Display the summary statistics
print(summary_stats)
```

```python
# Assuming the data is already loaded into a DataFrame called loyers1

# Calculate the difference between upper and lower bounds
loyers1['Plage de prix'] = loyers1['Limite supérieure de l\'IP par mètre carré'] - loyers1['Limite inférieure de l\'IP par mètre carré']

# Sort the DataFrame by price range in descending order
sorted_df = loyers1.sort_values(by='Plage de prix', ascending=False)

# Reset the index of the sorted DataFrame
sorted_df = sorted_df.reset_index(drop=True)

# Display the cities with the largest price range
cities_with_largest_range = sorted_df[['Localisation', 'Plage de prix']]
print(cities_with_largest_range)
```

```python
import matplotlib.pyplot as plt

# Select the relevant columns
rents = loyers1["Loyer prédit par mètre carré"]
lower_bound = loyers1["Limite inférieure de l'IP par mètre carré"]
upper_bound = loyers1["Limite supérieure de l'IP par mètre carré"]

# Create subplots for histograms
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Plot histograms
axs[0].hist(rents, bins=30, color="blue", alpha=0.5)
axs[1].hist(lower_bound, bins=30, color="green", alpha=0.5)
axs[2].hist(upper_bound, bins=30, color="orange", alpha=0.5)

# Set titles and labels
axs[0].set_title("Distribution of Predicted Rents per Square Meter")
axs[1].set_title("Distribution of Lower Bounds of IP per Square Meter")
axs[2].set_title("Distribution of Upper Bounds of IP per Square Meter")

axs[0].set_xlabel("Rent per Square Meter")
axs[1].set_xlabel("Lower Bound of IP per Square Meter")
axs[2].set_xlabel("Upper Bound of IP per Square Meter")

# Show the plots
plt.tight_layout()
plt.show()

# Create box plots
plt.figure(figsize=(8, 6))
plt.boxplot([rents, lower_bound, upper_bound], labels=["Predicted Rent", "Lower Bound of IP", "Upper Bound of IP"])
plt.title("Box Plot of Rent and Interval Predictions")
plt.ylabel("Price (per square meter)")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculer le prix moyen des loyers par région
prix_moyen_par_region = loyers1.groupby('Région')['Loyer prédit par mètre carré'].mean()

# Trier les régions par ordre décroissant des prix moyens
prix_moyen_par_region = prix_moyen_par_region.sort_values(ascending=False)

# Afficher les régions et leurs prix moyens dans un graphique de barres
plt.figure(figsize=(12, 6))
prix_moyen_par_region.plot(kind='bar')
plt.xlabel('Région')
plt.ylabel('Prix moyen des loyers')
plt.title('Régions les plus chères de France')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculer le prix moyen des loyers par département
prix_moyen_par_departement = loyers1.groupby('Département')['Loyer prédit par mètre carré'].mean()

# Trier les départements par ordre décroissant des prix moyens
prix_moyen_par_departement = prix_moyen_par_departement.sort_values(ascending=False)

# Sélectionner les 10 départements les plus chers
top_10_departements = prix_moyen_par_departement[:10]

# Afficher les départements et leurs prix moyens dans un graphique de barres
plt.figure(figsize=(12, 6))
top_10_departements.plot(kind='bar')
plt.xlabel('Département')
plt.ylabel('Prix moyen des loyers')
plt.title('Les 10 départements les plus chers de France')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculer le prix moyen des loyers par commune
prix_moyen_par_commune = loyers1.groupby('Localisation')['Loyer prédit par mètre carré'].mean()

# Trier les communes par ordre décroissant des prix moyens
prix_moyen_par_commune = prix_moyen_par_commune.sort_values(ascending=False)

# Afficher les communes et leurs prix moyens dans un graphique de barres (prendre les N communes les plus chères)
N = 10  # Nombre de communes à afficher
plt.figure(figsize=(12, 6))
prix_moyen_par_commune[:N].plot(kind='bar')
plt.xlabel('Localisation')
plt.ylabel('Prix moyen des loyers')
plt.title('Communes les plus chères de France')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Seuil pour les prix des loyers élevés (à définir selon vos critères)
seuil_prix_loyer = 15.0

# Calculer le prix moyen des loyers par commune
prix_moyen_par_commune = loyers1.groupby('Localisation')['Loyer prédit par mètre carré'].mean()

# Sélectionner les communes avec des prix de loyers élevés
communes_cheres = prix_moyen_par_commune[prix_moyen_par_commune > seuil_prix_loyer]

# Obtenir les régions correspondantes aux communes chères
regions_communes_cheres = loyers1[loyers1['Localisation'].isin(communes_cheres.index)]['Région'].unique()

# Afficher les régions correspondantes
print("Régions avec des communes ayant des prix de loyers élevés :")
for region in regions_communes_cheres:
    print(region)

# Visualisation des régions avec des communes ayant des prix de loyers élevés
loyers1[loyers1['Région'].isin(regions_communes_cheres)].groupby('Région')['Loyer prédit par mètre carré'].mean().plot(kind='bar')
plt.xlabel('Région')
plt.ylabel('Prix moyen des loyers')
plt.title('Régions avec des communes ayant des prix de loyers élevés')
plt.xticks(rotation=45)
plt.show()
```
