```python
# Proportion de personnes en état de sous-nutrition
```

```python
prop_sous_nutrition = pd.merge(population.loc[population['Année'] == 2017,["Zone", "Valeur"]],
                               sous_nutrition.loc[sous_nutrition['Année'] == '2016-2018',["Zone", "sous_nutrition"]],
                               on='Zone')
```

```python
prop_sous_nutrition
```

```python
Réponse1 = sum(prop_sous_nutrition['sous_nutrition']/sum(prop_sous_nutrition['Valeur'])*100)
```

```python
print("la proportion de personnes en état de sous-nutrition est de :", round(Réponse1,2), "%")
```

```python
dispo_alimentaire = dispo_alimentaire.merge(population.loc[population['Année'] == 2017, ["Zone", "Valeur"]], on='Zone')
```

```python
dispo_alimentaire['dispo_kcal'] = dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_alimentaire ['Valeur'] * 365 
```

```python
dispo_alimentaire_totale_en_kcal = dispo_alimentaire['dispo_kcal'].sum()
```

```python
dispo_alimentaire_totale_en_kcal
```

```python
potentiel_dhumains_nourrissables = round(dispo_alimentaire_totale_en_kcal/(2500*365))
```

```python
# Part des animaux, des pertes et de nourriture dans la disponibilité intérieure
```

```python
dispo_alimentaire
```

```python
dispo_alim_animaux = dispo_alimentaire["Aliments pour animaux"].sum()
print("La disponibilité intérieure pour les animaux est de :", round(dispo_alim_animaux), "")
```

```python
dispo_alim_nourriture = dispo_alimentaire["Nourriture"].sum()
print("La disponibilité intérieure pour la nourriture est de :", round(dispo_alim_nourriture))
```

```python
dispo_alim_pertes = dispo_alimentaire["Pertes"].sum()
print("La disponibilité intérieure perdue est de :", round(dispo_alim_pertes))
```

```python
dispo_alim_autres = dispo_alimentaire["Autres Utilisations"].sum()
print("La disponibilité intérieure autre est de :", round(dispo_alim_autres))
```

```python
dispo_alim_semences = dispo_alimentaire["Semences"].sum()
print("La disponibilité intérieure pour les semences est de :", round(dispo_alim_semences))
```

```python
dispo_interieure = dispo_alimentaire["Disponibilité intérieure"].sum()
print("La disponibilité intérieure totale est de :", round(dispo_interieure))
```

```python
part_animaux = (dispo_alim_animaux/dispo_interieure)*100
print("La part pour les animaux est de :", round(part_animaux, 2), "%") 
```

```python
part_nourriture = (dispo_alim_nourriture/dispo_interieure)*100
print("La part pour la nourriture des humains est de :", round(part_nourriture, 2), "%")
```

```python
part_pertes = (dispo_alim_pertes/dispo_interieure)*100
print("La part de perdue est de :", round(part_pertes, 2), "%")
```
